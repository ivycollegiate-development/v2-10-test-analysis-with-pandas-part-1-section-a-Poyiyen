```python
# Core data manipulation libraries
import pandas as pd  # For working with DataFrames
import numpy as np   # For numerical operations

# Machine learning tools from scikit-learn
from sklearn.model_selection import train_test_split  # For creating training/testing datasets
from sklearn.linear_model import LinearRegression     # Our basic prediction algorithm
from sklearn.metrics import mean_squared_error, r2_score  # For evaluating model performance

# Visualization
import matplotlib.pyplot as plt  # For creating plots
import seaborn as sns            # For enhanced visualizations

# Load the dataset
df = pd.read_csv('Pandas_Lesson_Grades.csv')
print(f"Loaded dataset with {df.shape[0]} rows and {df.shape[1]} columns")
```

```python
# Convert percentage strings to floats
# This removes the '%' symbol and converts to numeric values
df['% Correct'] = df['% Correct'].str.rstrip('%').astype(float)

# Let's examine the 'Responses' column to understand its structure
print("Example 'Responses' values:")
print(df['Responses'].head(3))

# This helper function extracts features from the response patterns
# It's provided for you - review it to understand what it does
def extract_response_features(response_str):
    """
    Extracts numerical features from response pattern strings like "A:69% B:23% C:8%"
    
    Parameters:
        response_str (str): The response pattern string
        
    Returns:
        dict: Dictionary containing extracted features
    """
    features = {}
    responses = response_str.split()
    
    # Calculate number of options used
    features['num_options'] = len(responses)
    
    # Calculate percentage for most popular wrong answer
    percentages = []
    for resp in responses:
        option, pct = resp.split(':')
        pct_value = float(pct.rstrip('%'))
        percentages.append(pct_value)
    
    # Get primary answer and its percentage
    primary_ans = df.loc[df['Responses'] == response_str, '# Pri. Ans.'].values[0]
    primary_pct = None
    
    for resp in responses:
        option, pct = resp.split(':')
        if option == primary_ans:
            primary_pct = float(pct.rstrip('%'))
            break
    
    features['primary_pct'] = primary_pct
    
    # Calculate response variance (spread of answers)
    features['response_variance'] = np.var(percentages)
    
    return features

# Apply function to each row and create new feature columns
# This creates new columns in our DataFrame
response_features = df['Responses'].apply(extract_response_features)
# Convert the dictionary results to DataFrame columns
extracted_df = pd.json_normalize(response_features)
# Join the new columns to our original DataFrame
df = pd.concat([df, extracted_df], axis=1)

# Display our new features
print("\nNewly created features:")
print(df[['num_options', 'primary_pct', 'response_variance']].head())
```

```python
# Create a correlation matrix for numeric columns
# First we need to identify which columns are numeric
numeric_cols = df.select_dtypes(include=['number']).columns.tolist()
print(f"Numeric columns: {numeric_cols}")

# Calculate correlation matrix
corr_matrix = df[numeric_cols].corr()

# Visualize correlation matrix with a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Correlation Matrix of Numeric Features')
plt.tight_layout()
plt.show()

# RESEARCH TASK: Look up how to create a scatterplot matrix using pandas or seaborn
# Hint: Search for "pandas scatter_matrix" or "seaborn pairplot"
# Your researched code here:


# Plot the relationship between Discrimination Factor and % Correct
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Discrim. Factor', y='% Correct')
plt.title('Relationship Between Discrimination Factor and Question Difficulty')
plt.xlabel('Discrimination Factor')
plt.ylabel('Percentage Correct (Easier )')

# Add regression line to see trend
sns.regplot(data=df, x='Discrim. Factor', y='% Correct', 
            scatter=False, ci=None, line_kws={"color":"red"})

plt.grid(True, alpha=0.3)
plt.show()
```

```python
# We'll create several new features that might help predict question difficulty

# 1. Create a feature for response distribution entropy (measures answer spread)
# Entropy measures how evenly distributed the responses are
# High entropy = answers spread evenly across options (could indicate guessing)
# Low entropy = answers concentrated on fewer options (could indicate clarity)

from scipy.stats import entropy

def calculate_entropy(response_str):
    """Calculate Shannon entropy of response distribution"""
    # Extract percentages from response string
    percentages = []
    for resp in response_str.split():
        pct = float(resp.split(':')[1].rstrip('%'))
        percentages.append(pct)
    
    # Convert to probabilities (divide by 100)
    probabilities = [p/100 for p in percentages]
    
    # Calculate entropy
    return entropy(probabilities, base=2)

# Apply to each row
df['entropy'] = df['Responses'].apply(calculate_entropy)
print(f"Entropy range: {df['entropy'].min():.2f} to {df['entropy'].max():.2f}")
```

```python
# This will let us test if earlier answer positions are easier/harder

# TASK: Write code to convert primary answer letters to positions
# A=1, B=2, C=3, D=4, etc.
# Your code here:
answer_positions = {'A': 1, 'B': 2, 'C': 3, 'D': 4, 'E': 5}
df['primary_pos'] = df['# Pri. Ans.'].map(answer_positions)
```

```python
# This ratio compares how many students got it right vs. wrong
df['correct_ratio'] = df['# Correct'] / df.shape[0]
```
